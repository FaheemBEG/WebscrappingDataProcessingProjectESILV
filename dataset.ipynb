```python
import re
# Fonction pour standardiser les modèles de GPU et CPU
def standardize_models(data_string, model_type):
    if pd.isna(data_string):
        return None
    pattern = ''
    if model_type == 'gpu':
        pattern = r'([A-Za-z]+\s*[A-Za-z]*\s*\d+[A-Za-z]*)'  # GPU pattern
    elif model_type == 'cpu':
        pattern = r'([A-Za-z]+\s*\d+\s*[A-Za-z]*\s*\d*[A-Za-z]*)'  # Modified CPU pattern
    extracted_models = re.findall(pattern, data_string)
    standardized_models = [model.replace(" ", "").upper() for model in extracted_models]
    return standardized_models

# Appliquer la fonction de standardisation
df_games_full_2['Standardized_GPU'] = df_games_full_2['GPU'].apply(lambda x: standardize_models(x, 'gpu'))
df_games_full_2['Standardized_CPU'] = df_games_full_2['CPU'].apply(lambda x: standardize_models(x, 'cpu'))

# Fonction pour trouver la meilleure correspondance
def find_best_match(standardized_models, model_df):
    if not standardized_models:
        return None
    model_matches = {model: 0 for model in model_df['Model']}
    for s_model in standardized_models:
        for model in model_df['Model']:
            if s_model in model.upper().replace(" ", ""):
                model_matches[model] += 1
    best_match = max(model_matches, key=model_matches.get)
    return best_match if model_matches[best_match] > 0 else None

# Recherche des meilleures correspondances
df_games_full_2['Matched_GPU'] = df_games_full_2['Standardized_GPU'].apply(lambda x: find_best_match(x, df_gpu))
df_games_full_2['Matched_CPU'] = df_games_full_2['Standardized_CPU'].apply(lambda x: find_best_match(x, df_processors))

# Jointure pour ajouter la valeur TDP du GPU
df_games_full_2 = df_games_full_2.merge(df_gpu, left_on='Matched_GPU', right_on='Model', how='left')
df_games_full_2.rename(columns={'TDP (Watts)': 'GPU_TDP'}, inplace=True)
df_games_full_2.drop('Model', axis=1, inplace=True)

# Jointure pour ajouter la valeur TDP du CPU
df_games_full_2 = df_games_full_2.merge(df_processors, left_on='Matched_CPU', right_on='Model', how='left')
df_games_full_2.rename(columns={'TDP (W)': 'CPU_TDP'}, inplace=True)
df_games_full_2.drop(['Model', 'Standardized_GPU', 'Matched_GPU', 'Standardized_CPU', 'Matched_CPU'], axis=1, inplace=True)

df_games_full_2.fillna(0, inplace=True)

df_games_full_2
```

```python
max_gpu_tdp = df_games_full_2['GPU_TDP'].max()
max_cpu_tdp = df_games_full_2['CPU_TDP'].max()

min_gpu_tdp = df_games_full_2['GPU_TDP'].min()
min_cpu_tdp = df_games_full_2['CPU_TDP'].min()

def calculate_carbon_impact(gpu_tdp, cpu_tdp):
    if gpu_tdp == 0 or cpu_tdp == 0:
        return -1
    # Calculer le score de consommation combinée
    combined_tdp = gpu_tdp + cpu_tdp
    # Calculer le nombre d'étoiles en fonction du TDP
    stars = 5 - (combined_tdp - min_gpu_tdp - min_cpu_tdp) / (max_gpu_tdp + max_cpu_tdp - min_gpu_tdp - min_cpu_tdp) * 5
    rounded_stars = round(stars * 2) / 2
    return min(max(rounded_stars, 0), 5)

df_games_full_2['impact_carbone'] = df_games_full_2.apply(lambda row: calculate_carbon_impact(row['GPU_TDP'], row['CPU_TDP']), axis=1)
df_games_full_2.drop(['GPU_TDP', 'CPU_TDP'], axis=1, inplace=True)

df_games_full_2
```
